```python
for i in range(10):
    test = DATA[i]
    left_hand_data =  test[:,:42]
    lips_data = test[: ,42:122]
    right_hand_data = test[:,122:]

    # Vẽ biểu đồ:
    plt.figure(figsize=(10, 6))

    plt.subplot(3, 1, 1)
    plt.imshow(left_hand_data.T, aspect='auto', cmap='viridis', origin='lower')
    plt.title('Left Hand')

    plt.subplot(3, 1, 2)
    plt.imshow(right_hand_data.T, aspect='auto', cmap='viridis', origin='lower')
    plt.title('Right Hand')

    plt.subplot(3, 1, 3)
    plt.imshow(lips_data.T, aspect='auto', cmap='viridis', origin='lower')
    plt.title('Lips')

    plt.xlabel('Frame Number')
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Giả sử 'test' là dữ liệu của bạn
test = np.random.rand(120, 164)

# Khởi tạo biểu đồ
fig, (ax1, ax2, ax3) = plt.subplots(3, 1)

im1 = ax1.imshow(np.array([]).reshape(-1, 42), aspect='auto', cmap='viridis', origin='lower')
ax1.set_title('Left Hand')

im2 = ax2.imshow(np.array([]).reshape(-1, 80), aspect='auto', cmap='viridis', origin='lower')
ax2.set_title('Right Hand')

im3 = ax3.imshow(np.array([]).reshape(-1, 42), aspect='auto', cmap='viridis', origin='lower')
ax3.set_title('Lips')

# Hàm cập nhật dữ liệu
def update(i):
    # Cập nhật dữ liệu cho biểu đồ
    im1.set_array(test[:i, :42].T)
    im2.set_array(test[:i, 42:122].T)
    im3.set_array(test[:i, 122:].T)

    # Điều chỉnh lại giới hạn trục y
    ax1.set_ylim(0, test[:i, :42].shape[1])
    ax2.set_ylim(0, test[:i, 42:122].shape[1])
    ax3.set_ylim(0, test[:i, 122:].shape[1])

# Tạo animation
ani = FuncAnimation(fig, update, frames=range(120), repeat=False)

plt.show()
# test = np.load('./test/start_up.npy')
```
